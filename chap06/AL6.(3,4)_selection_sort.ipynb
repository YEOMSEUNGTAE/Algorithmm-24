### 01.문제정의

##### 텍스트 문자열에서 패턴 문자열을 찾는 알고리즘

### 02.알고리즘 설명

##### 패턴'p'의 길이을 'm'으로 설정하고, 'NO_OF_CHARS'인 리스트 'tbl'을 'm'으로 초기화 마지막 문자를 제외한 각 문자에 대해 테이블 값 설정
##### 패턴이 일치하면 시작 인덱스를 일치하지 않으면 -1을 반환한다.

### 04. 코드 개요

##### 함수명 shift_table(pat) , search_horspool(T,P)
##### 입력값 텍스트T, 패턴 P
##### 반환값 패턴이 일치하면 시작 인덱스(왼쪽) 일치하지 않으면 -1을 반환한다.

### 05. 코드

```python
NO_OF_CHARS = 128

def shift_table(pat):
    m = len(pat)
    tbl = [m] * NO_OF_CHARS
    for i in range(m - 1):
        tbl[ord(pat[i])] = m - 1 - i
    return tbl
```

### 06. 테스트 코드

```python
NO_OF_CHARS = 128

def shift_table(pat):
    m = len(pat)
    tbl = [m] * NO_OF_CHARS
    for i in range(m - 1):
        tbl[ord(pat[i])] = m - 1 - i
    return tbl

def search_horspool(T, P):
    m = len(P)
    n = len(T)
    t = shift_table(P)
    i = m - 1
    
    while i < n:
        k = 0
        while k < m and P[m - 1 - k] == T[i - k]:
            k += 1
        if k == m:
            return i - m + 1
        else:
            i += t[ord(T[i])]
    
    return -1
print("패턴의 위치:", search_horspool("APPLEMANGOBANANAGRAPE", "BANANA"))
```
