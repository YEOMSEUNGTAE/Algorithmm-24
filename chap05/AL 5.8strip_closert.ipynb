### 1) 문제 정의

입력으로 n개의 점의 좌표로 이루어진 리스트 p가 주어진다. 모든 점들이 x좌표를 기준으로 오름차순으로 정렬되게 만들고 분할 정복 전략을 사용해서 문제를 구한다.

### 2) 알고리즘 설명

리스트 p를 두개의 부분 리스트로 분할하고 한다. 왼쪽과 오른쪽 분할된 점들의 개수가 3 이하인 경우는 억지기법으로 결과를 계산하고, 점들이 4개이면 분할 정복 기법을 적용한다. 
최근접쌍의 거리가 각각 계산이 되면 양쪽중에서도 가장 작은값을 반환한다.

### 4) 코드개요

함수명: def strip_closest(P, d):
입력값: P = [(2,6),(4,7),(4,1),(5,1),(3,2)]
출력값: print("띠 영역내에서 d보다 작은 최근접쌍의 거리 찾기:", strip_closest(P, len(P)))
설명: p영역의 점들의 리스트를 입력받고, sort함수와 람다함수를 이용해서 리스트의 항목 p인 튜플을 y축을 따라 오름차순으로 정리한다. pi와 pj의 거리가 d_min 이내일때까지 y값을 기준으로 가장 작은점 pi를 시적점으로 잡고, pi와 pj의 거리를 계산한다. 작은 값인 dij가 d_min보다 작으면 d_min을 갱신하고 j를 증가시키며 반복한다.

### 5) 코드

```python
import math

def distance(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2) #3장 113p

def strip_closest(P, d):
    n = len(P)
    d_min = d

    for i in range(n):
        j = i + 1
        while j < n and P[j][1] - P[i][1] < d_min: 
            dij = distance(P[i], P[j])
            if dij < d_min:
                d_min = dij
            j += 1
    return d_min

P = [(2,6),(4,7),(4,1),(5,1),(3,2)]
P.sort(key=lambda point: point[1])  
print("띠 영역내에서 d보다 작은 최근접쌍의 거리 찾기:", strip_closest(P, len(P)))
```
