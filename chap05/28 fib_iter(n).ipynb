```python
import time

def multMat(a, b):
    # Matrix multiplication of 2x2 matrices
    return [
        [a[0][0] * b[0][0] + a[0][1] * b[1][0], a[0][0] * b[0][1] + a[0][1] * b[1][1]],
        [a[1][0] * b[0][0] + a[1][1] * b[1][0], a[1][0] * b[0][1] + a[1][1] * b[1][1]]
    ]

def fib_iter(n):
    if n < 2:
        return n
    last = 0
    current = 1
    for _ in range(2, n + 1):
        last, current = current, current + last
    return current

def fib(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n - 1) + fib(n - 2)

def fib_mat(n):
    if n < 2:
        return n
    mat = [[1, 1], [1, 0]]
    result = powerMat(mat, n - 1)
    return result[0][0]

def powerMat(x, n):
    if n == 1:
        return x
    elif n % 2 == 0:
        half_power = powerMat(x, n // 2)
        return multMat(half_power, half_power)
    else:
        return multMat(x, powerMat(x, n - 1))

# Timing the functions
t1 = time.time()
for i in range(10):
    fib_iter(3)
t2 = time.time()
for i in range(10):
    fib(3)
t3 = time.time()
for i in range(10):
    fib_mat(3)
t4 = time.time()

print("fib_iter 시간 : ", t2 - t1)
print("fib 시간: ", t3 - t2)
print("fib_mat 시간: ", t4 - t3)
```
